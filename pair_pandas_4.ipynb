```python
#con condiciones de filtrado
condicion1 = csv_2016['categoría'] == 'Hardware'
condicion2 = csv_2016['categoría'] == 'Software'

hard_soft1 = csv_2016[condicion1 | condicion2]
hard_soft1.tail()
```

```python
#con método isin()
condiciones = ['Hardware', 'Software']

hard_soft2 = csv_2016[csv_2016['categoría'].isin(condiciones)]
hard_soft2.tail()
```

```python
patron1 = 'Solicitud'
patron2 = 'Problema'

solicitud = hard_soft2[hard_soft2["tipo"].str.contains(patron1, case = False)]
solicitud.head(2)
```

```python
print(f'Se han registrado {len(solicitud)} solicitudes')
```

```python
problema = hard_soft2[hard_soft2["tipo"].str.contains(patron2, case = False)]
problema.head(2)
```

```python
print(f'Se han registrado {len(problema)} problemas')
```

```python
#Solicitudes >3 satisfacción
satis_solicitud_3 = solicitud[solicitud["satisfacción"] > 3]
```

```python
print(f'Un total de {len(satis_solicitud_3)} solicitudes tienen una satisfacción de más de 3')
```

```python
#Solicitudes >4 satisfacción y <3 días
condicion_solic_1 = solicitud["satisfacción"] > 4
condicion_solic_2 = solicitud["días_resolución"] < 3

satis_solicitud_4 = solicitud[condicion_solic_1 & condicion_solic_2]
satis_solicitud_4.head()
```

```python
print(f'Un total de {len(satis_solicitud_4)} solicitudes tienen una satisfacción de más de 4 y menos de 3 días de resolución')
```

```python
mas_cinco_dias = csv_2016[csv_2016['días_resolución'] >= 5]
mas_cinco_dias.head()
```

```python
menos_cinco_dias = csv_2016[csv_2016['días_resolución'] < 5]
menos_cinco_dias.head()
```

```python
mas_cinco_dias['satisfacción'].unique()
```

```python
menos_cinco_dias['satisfacción'].unique()
```

```python
#En los dos casos tenemos puntuaciones de 1, 2, 3, 4 y 5
```

```python
mas_cinco_dias['satisfacción'].value_counts()
```

```python
menos_cinco_dias['satisfacción'].value_counts()
```

```python
'''A priori hay más puntuaciones de 5 en los que se han resuelto en menos de 5 días, pero también hay mayor número de casos en general que se han resuelto en menos de 5 días
(por ejemplo, también hay más puntuaciones de 1 en este df)'''
```

```python
mas_cinco_dias[['satisfacción']].describe().T
```

```python
menos_cinco_dias[['satisfacción']].describe().T
```

```python
'''Observando los estadísticos de cada df, parece que la media de satisfacción es ligeramente superior (pero prácticamente igual) cuando los tickets se resuelven en menos de 5 días, 
por lo que no parece que tenga una relevancia significativa'''
```

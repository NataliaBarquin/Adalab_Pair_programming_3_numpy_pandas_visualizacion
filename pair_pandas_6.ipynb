```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_empleados = pd.read_csv("data/puestos_empleados.csv")
df_empleados.head()
```

```python
df_tickets = pd.read_csv("data/union_años.csv")
df_tickets.head()
```

```python
df_empleados.rename(columns = {col :  col.lower().replace(" ", "_") for col in df_empleados.columns} , inplace = True)
df_empleados = df_empleados.drop(["unnamed:_0"], axis = 1)
df_empleados.head()
```

```python
df_tickets.rename(columns = {col :  col.lower().replace(" ", "_") for col in df_tickets.columns} , inplace = True)
df_tickets = df_tickets.drop(["unnamed:_0"], axis = 1)
df_tickets.head()
```

```python
concat1 = pd.concat([df_tickets, df_empleados], axis=0)
concat1
```

```python
#con condiciones de filtrado
condicion1 = concat1['categoría'] 

```

```python
condicion1
```

```python
df_tik = concat1.groupby("categoría")['id_ticket'].count()
df_tik
```

```python
categoria = pd.DataFrame(df_tik)
categoria.reset_index(inplace = True)
categoria
```

```python
categoria2 = categoria.sort_values("id_ticket", ascending = False)
categoria2
```

```python
df_media_satis = concat1.groupby(["categoría"])['satisfacción'].mean()
df_media_satis
```

```python
df_media_satis = concat1.groupby(["categoría"])['satisfacción'].agg(["mean", "std"]).reset_index()
df_media_satis
```

```python
df_comparacion = concat1.groupby(["categoría", "tipo"])["satisfacción"].mean()
df_comparacion
```

```python
df_empleado_de = concat1.groupby("departamento")['id_empleado'].count()
df_empleado_de
```

```python
empleados_depart = pd.DataFrame(df_empleado_de)
empleados_depart.reset_index(inplace = True)
empleados_depart
```

```python
emplea_depar = empleados_depart.sort_values("id_empleado", ascending = False)
emplea_depar
```

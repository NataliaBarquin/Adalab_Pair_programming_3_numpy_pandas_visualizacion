```python
import matplotlib.pyplot as plt
import pandas as pd
import re
import datetime
import numpy as np
```

```python
df_años = pd.read_csv("data/union_años.csv")
df_años.head()
```

```python
df_puestos = pd.read_csv("data/puestos_empleados.csv")
df_puestos.head()
```

```python
def sacar_mes(fecha):
    try:
        
        return fecha.split("-")[1]
    except:
        return np.nan
```

```python
df_años["meses"] = (df_años["fecha"].apply(sacar_mes))
```

```python
df_años
```

```python
def cambio_meses(meses):
  
  dic_fechas = {"01": "Enero", "02":"Febrero", "03":"Marzo","04":"Abril","05":"Mayo", "06":"Junio", "07": "Julio", "08": "Agosto", "09":"septiembre", "10":"Octubre", "11":"Noviembre","12":"Diciembre"}

  try:
    return dic_fechas[meses]
      
  except:
        
    return np.nan
```

```python
df_años["mes"] = (df_años["meses"].apply(cambio_meses))
```

```python
def unir_columnas(col1, col2):
    
    try:
      col3 = col1 + " " + col2
   
      return col3
    except:
        pass
```

```python
df_puestos["union_nombres"] = df_puestos.apply(lambda lola: unir_columnas(lola["Nombre"], lola["Apellido"]), axis = 1)
```

```python
def cambio_caracteres(categoría):
  
  dic_caract = {"Acceso&%Login": "Acceso/Login", 'Acceso99Login': "Acceso/Login",'Acceso-&Login':'Acceso/Login', 'Acceso%&Login': 'Acceso/Login', 'Hardware': 'Hardware', 'Software':'Software', 'Sistema':'Sistema'}

  try:
    return dic_caract[categoría]
      
  except:
        pass
```

```python
df_años["Categorías"] = (df_años["categoría"].apply(cambio_caracteres))
```

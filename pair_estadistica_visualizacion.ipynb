```python
"""
    - ¿ Hay una relación entre el azucar y la calidad?
    - ¿ Hay una relación entre el azucar y el nivel de alcohol?
    
    
"""
```

```python
plt.figure(figsize =(20,15))
fig = sns.swarmplot(x='quality', y = 'residual sugar' ,data = df_vino_blanco)
fig.set_title('Diagrama de enjambre calidad respecto al azucar ')
plt.show()
```

```python
plt.figure(figsize = (15,9))
ejercicio2 = sns.scatterplot(x = 'alcohol', y = 'residual sugar', data = df_vino_blanco, hue = 'quality')
ejercicio2.set(xlabel = 'alcohol_grados', ylabel = 'azucar')
ejercicio2.set_title('Relacion de azucar y alcohol respecto a la calidad del vino blanco')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```
